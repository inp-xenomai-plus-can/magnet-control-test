```python
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (16, 12)
from IPython import display
from matplotlib import pyplot as plt
```

```python
data = open('mes5.txt')

chans = {4:[],5:[],8:[],9:[]}
for line in data:
    nums = line.replace('\n','').split('\t')
    nums[0] = int(nums[0])
    nums[1] = float(nums[1])
    nums[2] = float(nums[2])
    if int(nums[0]) in chans.keys():
        chans[nums[0]].append(tuple(nums[1:]))
```

```python
# Calibrate
for i in range(2):
    cf = min(chans[4 + i])[0]/min(chans[8 + i])[0]
    print(cf)
    chans[8 + i] = [(v*cf, t) for v, t in chans[8 + i]]
```

```python
names = {4:'X1',5:'Y1',8:'X2',9:'Y2'}
for k, l in chans.items():
    px = []
    py = []
    for v, t in l:
        px.append(t)
        py.append(v)
    plt.plot(px, py, label=names[k])
plt.legend()
plt.xlabel('time, s')
plt.ylabel('adc input, V')
plt.show()
```

```python
names = ['X1 - X2', 'Y1 - Y2']
for i in range(2):
    pdv = []
    pt = []
    for (v1, t1), (v2, t2) in zip(chans[4 + i], chans[8 + i]):
        pdv.append(v1 - v2)
        pt.append(0.5*(t1 + t2))
    plt.plot(pt, pdv, label=names[i])
plt.legend()
plt.xlabel('time, s')
plt.ylabel('input diff, V')
plt.show()
```
